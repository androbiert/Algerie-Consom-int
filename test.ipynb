```python
from sklearn.cluster import KMeans
import numpy as np

# Example data
new_points = np.array([[1, 2], [1, 4], [1, 0],
                       [10, 2], [10, 4], [10, 0],
                       [5, 5], [6, 5], [5, 6]])

# Fit KMeans
model = KMeans(n_clusters=3, random_state=42)
model.fit(new_points)

# Get centroids
centroids = model.cluster_centers_
print(centroids)
centroids[:,0]
```

```python
from sklearn.decomposition import NMF
import numpy as np

# Imagine 4 documents, each with counts of 6 words
X = np.array([
    [1, 1, 0, 0, 2, 0],
    [2, 1, 0, 0, 3, 0],
    [0, 0, 2, 3, 0, 1],
    [0, 0, 3, 4, 0, 2]
])

# Apply NMF to find 2 hidden topics
nmf = NMF(n_components=2, random_state=42)
W = nmf.fit_transform(X)  # documents -> topics
H = nmf.components_       # topics -> words

print("Document-topic matrix (W):\n", W)
print("\nTopic-word matrix (H):\n", H)
```
